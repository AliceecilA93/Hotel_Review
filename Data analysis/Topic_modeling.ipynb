```python
from konlpy.tag import Mecab

def text_preprocess(doc):
  tokenizer = Mecab()
  clean_doc =  text_cleaning(doc)
  token_tagged = tokenizer.pos(clean_doc)
  nouns = [s for s, t in token_tagged if t in ['SL', 'NNG', 'NNP'] and len(s) > 1]
  morphs = [x for x in nouns if x not in stopwords] 
  preprocessed = ' '.join(morphs)
  return preprocessed
```

```python
from tqdm import tqdm
tqdm.pandas() 

df_review['abs_review'] = df_review['review'].progress_apply(lambda x: text_preprocess(x))
df_review
```

```python
def my_tokenizer(text):
  return text.split()
```

```python
def my_tokenizer_with_bigram(text):
    tokens = text.split()
    return [' '.join(tokens[i:i+2]) for i in range(len(tokens)-1)]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

전처리 된 리뷰를 통해 바이그램 돌리기

```python
tfidf_vect = TfidfVectorizer(tokenizer=my_tokenizer, ngram_range=(2, 2))
tfidf = tfidf_vect.fit_transform(raw_documents = df_review[df_review['청결'] == 1]['clean_review'])

tfidf
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components = 5, #토픽 갯수
                                max_iter = 10, #얼마나 돌린것인가
                                learning_method='online',
                                random_state=100)
lda_output = lda.fit_transform(tfidf)
```

```python
print('lda components:\n',lda.components_) 
print('shape:', lda.components_.shape)
```

```python
terms = tfidf_vect.get_feature_names() # 단어 집합

def get_topics(components, feature_names, n=7):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(2)) for i in topic.argsort()[:-n - 1:-1]])
get_topics(lda.components_,terms)
```
